```python
import scprep
import matplotlib as plt
```

```python

data_paths = [
    "../data/14_EAE_all_cells.MPS12344479-A02.sorted.1229.merged.10x_outputs/filtered_feature_bc_matrix",
    "../data/15_EAE_all_cells.MPS12344479-B02.sorted.1229.merged.10x_outputs/filtered_feature_bc_matrix",
    "../data/16_WT_all_cells.MPS12344479-H01.sorted.1229.merged.10x_outputs/filtered_feature_bc_matrix"
]

seurat_path = "../../tutorials/seurat_guidedClustering/seurat_clusteringTutorial/filtered_gene_bc_matrices/hg19"

data = scprep.io.load_10X(seurat_path, sparse=True, gene_labels='both')
# T1 = scprep.io.load_10X(data_paths[0], sparse=True, gene_labels='both')
# T2 = scprep.io.load_10X(data_paths[1], sparse=True, gene_labels='both')
# T3 = scprep.io.load_10X(data_paths[2], sparse=True, gene_labels='both')

# # Concatenating data matrices
# data, sample_labels = scprep.utils.combine_batches(
#     data=[T1, T2, T3],
#     batch_labels=["14_EAE", "15_EAE", "16_WT"],
#     append_to_cell_names=True
# )

# del T1, T2, T3
```

```python

data = scprep.filter.filter_empty_cells(data)
data = scprep.filter.filter_empty_genes(data)
data.head()
```

```python
scprep.plot.plot_library_size(data, cutoff=500)
data = scprep.filter.filter_library_size(data, cutoff=500)
plt.pyplot.show()
```

# Filter by mithocondrial expression to remove dead cells

```python
mt_genes = scprep.select.get_gene_set(data, starts_with="MT")
scprep.plot.plot_gene_set_expression(data, genes=mt_genes, percentile=90)
data = scprep.filter.filter_gene_set_expression(data, genes=mt_genes,
                                                percentile=90)

plt.pyplot.show()
```

# Library size normalize

```python
data = scprep.normalize.library_size_normalize(data)
```

# Square root transform

```python
# num = data._get_numeric_data()
# num[num < 0] = 0
data = scprep.transform.sqrt(data)
```
