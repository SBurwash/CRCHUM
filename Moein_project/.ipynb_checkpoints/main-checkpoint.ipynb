```python
# Load moein data
# Change default axis since data is inverted

data = scprep.io.load_cbsv("../data/moein_data/moein.data.csv", cell_axis="column")
metadata = pd.read_csv("../data/moein_data/moein.meta_data.csv")
```

```python
# Calculate phate of data

data_phate = phate.PHATE(t=25).fit_transform(data)
data_phate = pd.DataFrame(data_phate, index=data.index)

data_phate
```

```python
# Display PHATE for moein data

scprep.plot.scatter2d(data_phate,figsize=(12,12), cmap="Spectral", ticks=False, label_prefix="PHATE")
```
